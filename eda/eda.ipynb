change dtypes

```python
df['Selling_Price'] = df['Selling_Price'].astype('float16')
df['Present_Price'] = df['Present_Price'].astype('float16')
df['Year'] = df['Year'].astype('int16')
df['Driven_kms'] = df['Driven_kms'].astype('int32')
df['Fuel_Type'] = df['Fuel_Type'].astype('category')
df['Selling_type'] = df['Selling_type'].astype('category')
df['YeTransmissionr'] = df['Transmission'].astype('category')
df['Owner'] = df['Owner'].astype('category')
```

write numeric and categorical columns to list

```python
cat_columns = df.select_dtypes(include=['category']).columns.tolist()
num_columns = df.select_dtypes(include=['number']).columns.tolist()
print(cat_columns)
print(num_columns)
```
